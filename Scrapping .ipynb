# Trying to store data

```python

chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

wait = WebDriverWait(driver, 20)
actions = ActionChains(driver)

driver.get("https://www.issy.com")

try:
    lang_btn = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="block-issy-gtranslate"]/div/a')))
except Exception as e:
    print("Button not found:", e)

drop_down_ville = wait.until(EC.element_to_be_clickable((By.ID, "menu-ma-ville-btn")))
drop_down_ville.click()
time.sleep(1)

categories_ville = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "niveau-1")))
data = [] 

for cat in categories_ville:
    category_name = cat.text
    print(f"\nCategory: {category_name}")
    actions.move_to_element(cat).perform()
    time.sleep(1)

    try:
        sub_links = cat.find_elements(By.CSS_SELECTOR, ".niveau-2 a")
        subcategories = []
        for sub in sub_links:
            sub_name = sub.text.strip()
            sub_url = sub.get_attribute("href")
            subcategories.append({"name": sub_name, "url": sub_url, "content": ""})
            print(f"   ‚Ü≥ Subcategory: {sub_name} | URL: {sub_url}")
    except Exception:
        subcategories = []

    data.append({
        "category": category_name,
        "subcategories": subcategories
    })

with open("issy_my_ville_structure.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("\nData successfully saved to issy_my_ville_structure.json")
```

# extracting content from the links

```python
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

wait = WebDriverWait(driver, 15)
driver.get("https://www.issy.com/numerique")
time.sleep(2)

try:
    main_content = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'term-niveau-3')))
    for elem in main_content:
        print("Main Content:", elem.text)

    sub_contain = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "attached-nodes")))

    for sub_elem in sub_contain:
        links = sub_elem.find_elements(By.TAG_NAME, "a")

        for i, link in enumerate(links, start=1):
            href = link.get_attribute("href")
            text = link.text.strip()
            print(f" Visiting link {i}: {text} -> {href}")

            driver.get(href)

            try:
                content_elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "node__content")))
                for content in content_elements:
                    print(content.text)
            except Exception as e:
                print(" Could not extract node__content:", e)

            driver.back()
            time.sleep(2)

            sub_contain = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "attached-nodes")))
            sub_elem = sub_contain[0] 
            links = sub_elem.find_elements(By.TAG_NAME, "a")

except Exception as e:
    print("Could not extract text:", e)
```

# Extracting all links content at once

```python
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
wait = WebDriverWait(driver, 15)


with open("issy_links.txt", "r", encoding="utf-8") as f:
    category_links = [line.strip() for line in f if line.strip()]


for category_url in category_links:
    print(f"\nOpening category page: {category_url}")
    driver.get(category_url)
    time.sleep(2)

    try:
        main_content = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'term-niveau-3')))
        for elem in main_content:
            print("Main Content:", elem.text)

        sub_contain = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "attached-nodes")))

        for sub_elem in sub_contain:
            links = sub_elem.find_elements(By.TAG_NAME, "a")

            for i, link in enumerate(links, start=1):
                href = link.get_attribute("href")
                text = link.text.strip()
                print(f" Visiting link {i}: {text} -> {href}")

                driver.get(href)
                time.sleep(2)

                try:
                    content_elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "node__content")))
                    for content in content_elements:
                        print(content.text)
                except Exception as e:
                    print("Could not extract node__content:", e)

                driver.back()
                time.sleep(2)

                sub_contain = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "attached-nodes")))
                sub_elem = sub_contain[0]
                links = sub_elem.find_elements(By.TAG_NAME, "a")

    except Exception as e:
        print("Could not extract text for:", category_url, "| Error:", e)

driver.quit()
```

# trying to save in txt

```python
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
wait = WebDriverWait(driver, 15)

try:
    lang_btn = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="block-issy-gtranslate"]/div/a')))
except Exception as e:
    print("Button not found:", e)

with open("issy_links.txt", "r", encoding="utf-8") as f:
    category_links = [line.strip() for line in f if line.strip()]

with open("issy_output.txt", "w", encoding="utf-8") as output:
    for category_url in category_links:
        print(f"\nOpening category page: {category_url}")
        output.write(f"\n=== Category Page: {category_url} ===\n")
        driver.get(category_url)
        time.sleep(2)
        try:
            main_content = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'term-niveau-3')))
            for elem in main_content:
                print("Main Content:", elem.text)
                output.write(f"Main Content: {elem.text}\n")
            sub_contain = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "attached-nodes")))
            for sub_elem in sub_contain:
                links = sub_elem.find_elements(By.TAG_NAME, "a")
                for i, link in enumerate(links, start=1):
                    href = link.get_attribute("href")
                    text = link.text.strip()
                    print(f" Visiting link {i}: {text} -> {href}")
                    output.write(f"\nVisiting link {i}: {text} -> {href}\n")
                    driver.get(href)
                    time.sleep(2)
                    try:
                        content_elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "node__content")))
                        for content in content_elements:
                            content_text = content.text.strip()
                            if content_text:
                                print(content_text)
                                output.write(content_text + "\n\n")
                    except Exception as e:
                        print("Could not extract node__content:", e)
                        output.write(f"Could not extract node__content: {e}\n")
                    driver.back()
                    time.sleep(2)
                    sub_contain = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "attached-nodes")))
                    sub_elem = sub_contain[0]
                    links = sub_elem.find_elements(By.TAG_NAME, "a")
        except Exception as e:
            print("Could not extract text for:", category_url, "| Error:", e)
            output.write(f"Could not extract text for {category_url} | Error: {e}\n")

driver.quit()
```
